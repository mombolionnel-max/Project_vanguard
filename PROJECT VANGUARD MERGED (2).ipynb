Time Spent on Each Step

Calculation of Time Spent on Each Step in GROUPS TEST AND CONTROL

```python
# conversion to datetime format for time calculations
merged['date_time'] = pd.to_datetime(merged['date_time'])

# Sort by visit_id and date_time to do calculation on the same visit session

merged = merged.sort_values(by=['visit_id', 'date_time'])

merged.head()
```

```python
# For each session, we calculate the difference between the current time and the time of the previous step (in a new column).
merged['time_diff'] = merged.groupby('visit_id')['date_time'].diff()
merged.head()
```

```python
# For each session, we calculate the step duration (in a new column).
merged['step_duration'] = merged.groupby('visit_id')['time_diff'].shift(-1)

merged.head()
```

```python


# Average time spent on each step by variation
average_time_spent = merged.groupby(['variation', 'process_step'])['step_duration'].mean()
average_time_spent_seconds = average_time_spent.dt.total_seconds().round(2)

#  step_order
step_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

#  indexation of columns 
merged_time = average_time_spent_seconds.reset_index()

#  Convert the 'process_step' column into an ordered categorical type
merged_time['process_step'] = pd.Categorical(
    merged_time['process_step'], 
    categories=step_order, 
    ordered=True)

# Sort the DataFrame by 'variation',then by the ordered order of 'process_step'
merged_time_sorted = merged_time.sort_values(by=['variation', 'process_step'])

merged_time_sorted


```

```python
import pandas as pd
import numpy as np


def remove_outliers_by_group(df, group_cols, time_col_timedelta):
    """
    Removes outliers from a time column using the 1.5 * IQR rule, 
    grouped by specified categorical columns.
    """
    # Convert timedelta to seconds for IQR calculation
    df['step_duration_sec'] = df[time_col_timedelta].dt.total_seconds()
    
    # Calculate Q1, Q3, and IQR for each group (variation and process_step)
    q1 = df.groupby(group_cols)['step_duration_sec'].transform(lambda x: x.quantile(0.25))
    q3 = df.groupby(group_cols)['step_duration_sec'].transform(lambda x: x.quantile(0.75))
    iqr = q3 - q1
    
    # Define the lower and upper bounds
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Filter out the outliers
    df_filtered = df[
        (df['step_duration_sec'] >= lower_bound) & 
        (df['step_duration_sec'] <= upper_bound)
    ].copy()
    
    # Clean up and return the filtered DataFrame
    return df_filtered.drop(columns=['step_duration_sec'])

#'merged' DataFrame and 'step_duration' column already exist.

# Apply the outlier removal to a copy of the merged data
merged_cleaned = remove_outliers_by_group(
    merged.copy(), 
    group_cols=['variation', 'process_step'], 
    time_col_timedelta='step_duration'
)

# Recalculate Average Time Spent on Each Step using the cleaned data
average_time_spent_cleaned = merged_cleaned.groupby(['variation', 'process_step'])['step_duration'].mean()

# Convert to seconds and format the breakdown table
step_time_df_cleaned = average_time_spent_cleaned.dt.total_seconds().round(2).unstack(level=0)

# Calculate the new Average Total Step Time (mean of all step averages)
avg_total_time_control = step_time_df_cleaned['Control'].mean().round(3)
avg_total_time_test = step_time_df_cleaned['Test'].mean().round(3)


# We will use the error calculation .
total_steps_test = 176641
total_steps_control = 140482
total_errors_test = 11589
total_errors_control = 5331

# Total unique visitors (used for % Users with Errors - this is an assumption based on completion rate data)
N_start_test = 33144
N_start_control = 30903

# Final Summary Data Construction
summary_data = {
    'KPI': [
        'Completion Rate (%)', 
        'Average Total Step Time (seconds)', 
        'Error Rate per Step (%)'
    ],
    'Control': [
        (16039 / 30903) * 100, 
        avg_total_time_control,
        (5331 / 140482) * 100
    ],
    'Test': [
        (21725 / 33144) * 100,
        avg_total_time_test,
        (11589 / 176641) * 100
    ]
}
summary_df = pd.DataFrame(summary_data).set_index('KPI').round(2)
summary_df.columns = ['Old Design', 'New Design']


print("--- Cleaned Step-by-Step Time Breakdown (Outliers Removed) ---")
print(step_time_df_cleaned.to_markdown(numalign="left", stralign="left"))

print("\n--- Final KPI Summary Table (Using Cleaned Time & Confirmed Rates) ---")
print(summary_df.to_markdown(numalign="left", stralign="left"))
```

Error Rates KPI

```python
merged_cleaned['date_time'] = pd.to_datetime(merged_cleaned['date_time'])

# define the numerical order of the process steps
step_order_merged = {
    'start': 1,
    'step_1': 2,
    'step_2': 3,
    'confirm': 4
}
# New column with step order
merged_cleaned['step_order_merged'] = merged_cleaned['process_step'].map(step_order_merged)
# Sort the data by session and by time
merged_cleaned= merged_cleaned.sort_values(by=['visitor_id', 'visit_id', 'date_time']).reset_index(drop=True)

#Identify the order of the next step for each line in the session
merged_cleaned['next_step_merged_order'] = merged_cleaned.groupby(['visitor_id', 'visit_id'])['step_order_merged'].shift(-1)
#Mark 'Backward' Steps (Errors) / An error occurs when: Current Step Order > Next Step Order
merged_cleaned['error'] = np.where(merged_cleaned['step_order_merged'] > merged_cleaned['next_step_merged_order'], 1, 0)

merged_cleaned.head()

```

```python
#Calculate the total errors and the total steps taken per 'variation' group
error_rate_kpi = merged_cleaned.groupby('variation').agg(
    total_backward_steps=('error', 'sum'),
    total_steps_taken=('client_id', 'count') )

error_rate_kpi
```

```python
# Calculer le Taux d'Erreur (%)
error_rate_kpi['Error_Rate'] = (error_rate_kpi['total_backward_steps'] / error_rate_kpi['total_steps_taken']) * 100

error_rate_kpi
```
